```python
num1=int(input("Enter number 01: "))
num2=int(input("Enter number 02: "))
num3=int(input("Enter number 03: "))
num4=int(input("Enter number 04: "))
num5=int(input("Enter number 05: "))
listdata=[num1,num2,num3,num4,num5]
for var in listdata:
    print(var)
    total=0
for var in listdata:
    total+=var
print("Sum:", total)


maxval = listdata[0]
minval = listdata[0]

for var in listdata:
    if var > maxval:
        maxval = var
    if var < minval:
        minval = var

print("Maximum value:", maxval)
print("Minimum value:", minval)

index = int(input("\nEnter index to update: "))
newvalue = int(input("Enter new value: "))
listdata[index] = newvalue

print("Updated list:", listdata)
```
