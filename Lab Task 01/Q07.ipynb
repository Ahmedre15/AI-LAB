```python
def repeatednumbers(lst):
    result = []

    for i in range(len(lst)):
        count = 0

        for j in range(len(lst)):
            if lst[i] == lst[j]:
                count += 1

        if count > 1 and lst[i] not in result:
            result.append(lst[i])

    return result

numbers = [1, 2, 3, 2, 4, 5, 1, 6, 3]
print(repeatednumbers(numbers))
```
