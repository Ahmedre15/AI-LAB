```python
text = input("Enter a string: ")

currentsubstring = ""
longestsubstring = ""

for ch in text:
    if ch not in currentsubstring:
        currentsubstring = currentsubstring + ch
    else:
        index = currentsubstring.index(ch)
        currentsubstring = currentsubstring[index + 1:] + ch

    if len(currentsubstring) > len(longestsubstring):
        longestsubstring = currentsubstring

print("Longest substring without repeating characters:", longestsubstring)
```
