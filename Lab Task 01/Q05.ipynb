```python
def strings(text):
    vowels = "aeiou"
    totalchars = 0
    vowelcount = 0
    consonantcount = 0

    for ch in text:
        if ch!=' ':         
            totalchars += 1
            if ch in vowels:
                vowelcount += 1
            else:
                consonantcount += 1

    return {
        "Total Characters": totalchars,
        "Total Vowels": vowelcount,
        "Total Consonants": consonantcount
    }

string = input("Enter a string: ")
result = strings(string)

print(result)
```
