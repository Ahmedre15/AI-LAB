```python
class Student:
    def __init__(self,name,):
        self.name=name
        self.__grade=0
    def setgrade(self,grade):
        if 0<=grade<=100:
            self.__grade=grade
    def getgrade(self):
        return self.__grade
    def display_info(self):
     print("Student Name:", self.name)
     print("Grade:", self.__grade)
        
s1 = Student("Ali")
s2 = Student("Sara")

s1.setgrade(85)
s2.setgrade(92)

print("Ali's Grade:", s1.getgrade())
print("Sara's Grade:", s2.getgrade())

s1.display_info()
s2.display_info()
```
