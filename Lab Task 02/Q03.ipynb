```python
class ResponseAgent:
    def response(self):
        print("Executing general response")
class AlertAgent(ResponseAgent):
    def response(self):
        print("Sending security alert notifications to admin.")
class BlockAgent(ResponseAgent):
    def response(self):
        print("Blocking malicious IP and terminating suspicious processes.")
class RecoverAgent(ResponseAgent):
    def response(self):
        print("Restoring affected systems and recovering data.")
agent1 = AlertAgent()
agent2 = BlockAgent()
agent3 = RecoverAgent()
agents = [agent1, agent2, agent3]
for agent in agents:
    agent.response()
```
