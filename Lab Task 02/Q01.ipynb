```python
class Securityagent:
  def __init__(self,agentid, name):
    self.agentid=agentid
    self.name=name
    self.status="Active"
  def showdetials(self):
    print("Agent Id: ",self.agentid)
    print("Name: ",self.name)
    print("Status: ",self.status)

class Firewallagent(Securityagent):
    def monitortraffic(self):
        print(self.name," is monitoring traffic.")
class Malwaredetectagent(Securityagent):
    def scanmalware(self):
        print(self.name," is scanning for malware.")        
class Automationagent(Securityagent):
    def runautomation(self):
        print(self.name," is running automation.")

firewall = Firewallagent(101, "Firewall Defender")
malware = Malwaredetectagent(102, "Malware Scanner")
automation = Automationagent(103, "Security Checker")


firewall.showdetials()
firewall.monitortraffic()

malware.showdetials()
malware.scanmalware()

automation.showdetials()
automation.runautomation()
```
