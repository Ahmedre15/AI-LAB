```python
class Threat:
    def __init__(self,id,name,severity):
        self.threatid=id
        self.name=name
        self.severity=severity
        self.status="Active"

    def showinfo(self):
        print("Threat ID: ",self.threatid)
        print("Name: ",self.name)
        print("Severity: ",self.severity)
        print("Status: ",self.status)
    
    def response(self):
        print("Generating threat reponse")

class Phishingthreat(Threat):
    def analyzingemail(self):
        print(self.name," is analyzing emails")

class RansomwareThreat(Threat):
    def scanfiles(self):
        print(self.name, "is scanning file system for encryption activity.")
class BotnetThreat(Threat):
    def detecttraffic(self):
        print(self.name, "is monitoring abnormal network traffic.")
phish = Phishingthreat(201, "Email Phishing", "High")
ransom = RansomwareThreat(202, "File Encryptor", "Critical")
bot = BotnetThreat(203, "Alien Network", "Medium")
phish.showinfo()
phish.analyzingemail()
phish.response()

ransom.showinfo()
ransom.scanfiles()
ransom.response()

bot.showinfo()
bot.detecttraffic()
bot.response()
```
