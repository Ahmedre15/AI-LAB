```python
class Employee:
    def work(self):
        print("Employee is working on general tasks.")


class Manager(Employee):
    def work(self):
        print("Manager is planning projects and managing team.")


class Developer(Employee):
    def work(self):
        print("Developer is writing and debugging code.")


class Designer(Employee):
    def work(self):
        print("Designer is creating UI/UX designs.")


employees = [Manager(), Developer(), Designer()]

for emp in employees:
    emp.work()
```
