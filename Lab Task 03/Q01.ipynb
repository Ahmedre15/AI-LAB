```python
import random

class SecuritySystem:
    def __init__(self):
        self.components = {}
        for comp in ["A","B","C","D","E","F","G","H","I"]:
            self.components[comp] = random.choice(["Safe","Vulnerable"])

    def display_system(self):
        print("System Status:")
        for comp, status in self.components.items():
            print("Component", comp, ":", status)


class SecurityAgent:
    def __init__(self):
        self.vulnerable_list = []

    def scan_system(self, system):
        print("\nScanning system..")
        for comp, status in system.components.items():
            if status == "Vulnerable":
                print("Warning:", comp, "is Vulnerable")
                self.vulnerable_list.append(comp)
            else:
                print(comp, "is Safe")

    def patch_system(self, system):
        print("\nPatching Vulnerabilities.")
        for comp in self.vulnerable_list:
            system.components[comp] = "Safe"
            print("Patched Component", comp)

        self.vulnerable_list.clear()


system = SecuritySystem()
agent = SecurityAgent()


system.display_system()

agent.scan_system(system)
agent.patch_system(system)


system.display_system()
```
