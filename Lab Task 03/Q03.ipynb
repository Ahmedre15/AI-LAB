```python
import random

class BackupSystem:
    def __init__(self):
        self.tasks = {}
        for task in range(1, 6):
            self.tasks[task] = random.choice(["Completed", "Failed"])

    def display_system(self):
        print("Backup Status:")
        for task, status in self.tasks.items():
            print("Task", task, ":", status)


class BackupManagementAgent:
    def __init__(self):
        self.failed = []

    def scan_system(self, system):
        print("\nScanning System...")

        for task, status in system.tasks.items():
            if status == "Failed":
                print("Task", task, "has Failed")
                self.failed.append(task)
            else:
                print("Task", task, "is Completed")

    def retry_system(self, system):
        print("\nRetrying Failed Tasks...")

        while self.failed:
            task = self.failed.pop()
            system.tasks[task] = "Completed"
            print("Task", task, "is now Completed")

        print("Retry Process Complete.")


system = BackupSystem()
agent = BackupManagementAgent()

print("INITIAL STATUS")
system.display_system()

agent.scan_system(system)
agent.retry_system(system)

print("\nFINAL STATUS")
system.display_system()
```
