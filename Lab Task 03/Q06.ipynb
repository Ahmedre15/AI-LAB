```python

class Building:
    def __init__(self):
        
        self.rooms = {
            'a': "safe", 'b': "safe", 'c': "fire",
            'd': "safe", 'e': "fire", 'f': "safe",
            'g': "safe", 'h': "safe", 'j': "fire"
        }

    def display_environment(self):
        print("\nBuilding Status:")

        grid = ['a','b','c','d','e','f','g','h','j']
        count = 0

        for room in grid:
            if self.rooms[room] == "fire":
                print("F", end=" ")  
            else:
                print(".", end=" ")

            count += 1
            if count % 3 == 0:
                print() 
class FireRobot:
    def __init__(self):
        self.path = ['a','b','c','d','e','f','g','h','j']
        self.current_room = None

    def move(self, building):
        print("Starting Firefighting Robot...\n")

        for room in self.path:
            self.current_room = room
            print("Robot moved to room", room)

            if building.rooms[room] == "fire":
                print("Fire detected in room", room, "→ Extinguishing...")
                building.rooms[room] = "safe"
            else:
                print("Room", room, "is safe.")


            building.display_environment()

        print("\nAll rooms checked. Mission Complete.")


building = Building()
robot = FireRobot()

print("=== INITIAL BUILDING STATUS ===")
building.display_environment()

robot.move(building)

print("\n=== FINAL BUILDING STATUS ===")
building.display_environment()
```
