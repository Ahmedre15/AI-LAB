```python
import random

class SecuritySystem:
    def __init__(self):
        self.components = {}

        for comp in ["A","B","C","D","E","F","G","H","I"]:
            self.components[comp] = random.choice(
                ["Safe", "Low Risk Vulnerable", "High Risk Vulnerable"]
            )

    def display_system(self):
        print("\nSystem Status:")
        for comp, status in self.components.items():
            print("Component", comp, ":", status)


class UtilitySecurityAgent:
    def __init__(self):
        self.low_risk = []
        self.high_risk = []
        self.utility = 0

    def scan_system(self, system):
        print("\nScanning System...")

        for comp, status in system.components.items():

            if status == "Low Risk Vulnerable":
                print("Warning:", comp, "has LOW Risk Vulnerability")
                self.low_risk.append(comp)

            elif status == "High Risk Vulnerable":
                print("Critical:", comp, "has HIGH Risk Vulnerability")
                self.high_risk.append(comp)

            else:
                print(comp, "is Safe")

    def patch_system(self, system):
        print("\nPatching Low Risk Vulnerabilities...")

        while self.low_risk:
            comp = self.low_risk.pop()
            system.components[comp] = "Safe"
            self.utility += 10  
            print("Patched Component", comp)

        if self.high_risk:
            print("\nHigh Risk Vulnerabilities require Premium Service.")
            for comp in self.high_risk:
                print("Component", comp, "remains High Risk.")

        print("\nTotal Utility Gained:", self.utility)


system = SecuritySystem()
agent = UtilitySecurityAgent()

print("INITIAL SYSTEM CHECK")
system.display_system()

agent.scan_system(system)
agent.patch_system(system)

print("\nFINAL SYSTEM STATUS")
system.display_system()
```
