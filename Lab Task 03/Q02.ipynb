```python
import random

class SecuritySystem:
    def __init__(self):
        self.servers = {}
        for server in range(1,6):
            self.servers[server] = random.choice(["Underloaded", "Balanced", "Overloaded"])

    def display_system(self):
        print("System Status:")
        for server, load in self.servers.items():
            print( server, ":", load)

class LoadBalancerAgent:
    def __init__(self):
        self.overload = []
        self.underload=[]
    def scan_system(self, system):
        print("\nScanning system..")
        for server, load in system.servers.items():
            if load == "Overloaded":
                print(server, "is Overloaded")
                self.overload.append(server)
            elif load == "Underloaded":
                print(server, "is Underloaded")
                self.underload.append(server)
            else:
                print(server, "is Balanced")
    def balancesystem(self, system):
        print("\nBalancing Loads...")
        while self.overload and self.underload:
            over = self.overload.pop()
            under = self.underload.pop()

            system.servers[over] = "Balanced"
            system.servers[under] = "Balanced"

            print("Moved tasks from", over, "to", under)

        print("Load Balancing Complete.")
system = SecuritySystem()
agent = LoadBalancerAgent()

print("=== INITIAL STATUS ===")
system.display_system()

agent.scan_system(system)
agent.balancesystem(system)

print("\n=== FINAL STATUS ===")
system.display_system()
```
