```python

class Hospital:
    def __init__(self):
        self.patients = {
            101: "Panadol",
            102: "Insulin",
            103: "Antibiotic"
        }

        self.staff_available = True

    def display_schedule(self):
        print("\nToday's Medicine Schedule:")
        for room, medicine in self.patients.items():
            print("Room", room, "needs", medicine)


class DeliveryRobot:
    def __init__(self):
        self.current_location = "Corridor"
        self.carrying_medicine = None
        self.goal = None

    def move_to(self, location):
        self.current_location = location
        print("Robot moved to", location)

    def pick_medicine(self, medicine):
        self.carrying_medicine = medicine
        print("Picked up", medicine)

    def scan_patient(self, room):
        print("Scanning patient ID in Room", room)
        return True  

    def deliver_medicine(self, room):
        print("Delivered", self.carrying_medicine, "to Room", room)
        self.carrying_medicine = None

    def alert_staff(self):
        print("Alerting medical staff!")

    def achieve_goal(self, hospital):
        print("\nStarting Delivery Process...")

        for room, medicine in hospital.patients.items():

            self.goal = "Deliver " + medicine + " to Room " + str(room)
            print("\nCurrent Goal:", self.goal)

            self.move_to("Medicine Storage")

            self.pick_medicine(medicine)

            self.move_to("Room " + str(room))

            verified = self.scan_patient(room)

            if verified:
                self.deliver_medicine(room)
            else:
               
                self.alert_staff()

        print("\nAll scheduled deliveries completed.")

hospital = Hospital()
robot = DeliveryRobot()

print("=== HOSPITAL SYSTEM ===")
hospital.display_schedule()

robot.achieve_goal(hospital)
```
