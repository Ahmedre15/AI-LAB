```python
class Environment:
    def __init__(self,heat_leavel='High'):
        self.heat_level=heat_leavel
    def get_percept(self):
        if self.heat_level=='High':
           return 'Hot'
        else:
            return 'Cool'
    def cool_down(self):
        self.heat_level = 'Low'    
class simplereflexagent:
    def __init__(self):
        pass
    def act(self,percept):
        if percept=='Hot':
           
            return 'Pull Hands away'
        
        else :
            return 'No need to pull away'
        
def runagent(agent,env,steps):
    for step in range(steps):
      percept=env.get_percept()
      action=agent.act(percept)
      print(f"Percept {step+1}: {percept}, Action: {action}")
      if percept=='Hot':
          environment.cool_down()

agent=simplereflexagent()
environment=Environment('High')
runagent(agent,environment,5)
```
