```python
import heapq
graph = {
    'S': [('A', 3), ('B', 6), ('C', 5)],
    'A': [('D', 9), ('E', 8)],
    'B': [('F', 12), ('G', 14)],
    'C': [('H', 7)],
    'H': [('I', 5), ('J', 6)],
    'I': [('K', 1), ('L', 10), ('M', 2)],
    'D': [],'E': [],
    'F': [],'G': [],
    'J': [],'K': [],
    'L': [],'M': []
}
def beamsearch(start,goal,beamwidth=2):
    beam=[(0,[start])]
    while beam:
        candidates=[]
        for cost,path in beam:
            current_node=path[-1]
            if current_node==goal:
                return cost,path
            for neighhbour,edge_cost in graph.get(current_node,[]):
               newcost=cost+edge_cost
               newpath=path+[neighhbour]
               candidates.append((newcost,newpath))
        beam=heapq.nsmallest(beamwidth,candidates,key=lambda x:x[10])
    return None,float('inf')
startnode='S'
endnode='L'
beamwidth=2
beamsearch(startnode,endnode,beamwidth=2)
```
