```python
#Question 6
# An AI agent is placed inside an unknown maze. The agent can only sense its current position and the
# immediate moves available from that position (up, down, left, or right).
# The location of the goal is not known in advance and may lie deep inside the maze. The maze can also
# contain dead ends, and the agent does not have any heuristic or distance information to guide its search.
# 1. Write Python code to implement:

# • Depth-First Search with a depth limit
# • Iterative Deepening Search

# Important constraints:
# • Do not use built-in stack or queue libraries for the main search logic.
# • You may use basic Python lists to manage nodes.


# ----------- Helper Functions -----------

def find_start(maze):
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 'S':
                return (i, j)

def find_goal(maze):
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if maze[i][j] == 'G':
                return (i, j)

def is_valid(maze, x, y, visited):
    rows = len(maze)
    cols = len(maze[0])
    return (0 <= x < rows and
            0 <= y < cols and
            maze[x][y] != 1 and
            (x, y) not in visited)

# ----------- Depth Limited DFS -----------

def depth_limited_dfs(maze, current, goal, limit, visited, path):
    if current == goal:
        return path

    if limit <= 0:
        return None

    visited.append(current)

    x, y = current
    moves = [(-1,0),(1,0),(0,-1),(0,1)]  # Up, Down, Left, Right

    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if is_valid(maze, nx, ny, visited):
            result = depth_limited_dfs(
                maze,
                (nx, ny),
                goal,
                limit - 1,
                visited,
                path + [(nx, ny)]
            )
            if result:
                return result

    visited.pop()
    return None

# ----------- Iterative Deepening Search -----------

def iterative_deepening_search(maze):
    start = find_start(maze)
    goal = find_goal(maze)

    depth = 0
    while True:
        visited = []
        result = depth_limited_dfs(
            maze,
            start,
            goal,
            depth,
            visited,
            [start]
        )
        if result:
            return result
        depth += 1

# 2. Code to test your implementations on the following sample maze, represented as a 2D grid and
# show the output result:
# maze = [
# ['S', 0, 0, 1, 0],
# [1, 0, 1, 0, 0],
# [0, 0, 0, 0, 'G'],
# [1, 1, 0, 1, 1]
# ]
# Where:
# • S = Start position
# • G = Goal position
# • 0 = Free cell
# • 1 = Obstacle


maze = [
['S', 0, 0, 1, 0],
[1, 0, 1, 0, 0],
[0, 0, 0, 0, 'G'],
[1, 1, 0, 1, 1]
]

start = find_start(maze)
goal = find_goal(maze)

print("Start:", start)
print("Goal:", goal)

# Depth Limited DFS
limit = 10
visited = []
dls_result = depth_limited_dfs(maze, start, goal, limit, visited, [start])
print("\nDepth-Limited DFS Result:")
print(dls_result)

# Iterative Deepening Search
ids_result = iterative_deepening_search(maze)
print("\nIterative Deepening Search Result:")
print(ids_result)
```

```python
# 3. Based on the behavior of both algorithms in this maze, explain analytically why Iterative
# Deepening Search is more suitable than Depth-First Search for this problem.

Iterative Deepening Search is more suitable for this problem because:

1.The goal location is unknown, and DFS may explore deep wrong paths before finding the goal.

2.The maze contains dead ends. DFS can go deep into dead ends and waste time.

3.IDS performs depth-limited searches with gradually increasing depth, ensuring that shallow goals are found first.

4.IDS is complete and optimal (for equal step cost), while DFS is not guaranteed to find the shortest path.

5.IDS uses the same low memory as DFS but behaves like BFS in terms of completeness.
```
